```python
import sympy as sp
import numpy as np
from ase import eos
from ase.units import eV, J, kJ
```

```python
Hex = np.transpose(np.loadtxt("hex.txt"))
Cub = np.transpose(np.loadtxt("Cub.txt"))


print(Cub)


EOScub=eos.EquationOfState(Cub[1], Cub[0]*J/eV, eos='birchmurnaghan')
EOScub.plot(filename='Cub.png')
v0, e0, B = EOScub.fit()
print('cub: ', B / kJ * 1.0e24, 'GPa')

```

```python
Hex = np.transpose(np.loadtxt("hex.txt"))
print(Hex)
EOShex=eos.EquationOfState(Hex[1], Hex[0]*J/eV, eos='birchmurnaghan')
EOShex.plot(filename='Hex.png')
v0, e0, B = EOShex.fit()
print('hex: ', B / kJ * 1.0e24, 'GPa')
```
